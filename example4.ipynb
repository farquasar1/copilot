```python
# get the data and read in into a pandas dataframe
import pandas as pd
data = pd.read_csv('https://raw.githubusercontent.com/alanjones2/dataviz/master/london2018.csv')
```

```python
data.head()
```

```python
# plot the data
import matplotlib.pyplot as plt
plt.plot(data['Tmax'])  
plt.plot(data['Tmin'])
plt.plot(data['Rain'])
plt.legend(['Max Temp', 'Min Temp', 'Rain'])
plt.show()
```

```python
# fill between min and max temp
plt.plot(data['Tmax'])
plt.plot(data['Tmin'])
plt.fill_between(range(365), data['Tmin'], data['Tmax'], alpha=0.5)
plt.show()
```

```python
# repeat las cell is has just 13 months of data
plt.plot(data['Tmax'])
plt.plot(data['Tmin'])
plt.fill_between(range(12), data['Tmin'], data['Tmax'], alpha=0.5)
plt.show()
```

```python
# use legends to label the plot and the axis. Use the names of the months 
# as the x-axis labels
plt.plot(data['Tmax'])
plt.plot(data['Tmin'])
plt.fill_between(range(12), data['Tmin'], data['Tmax'], alpha=0.5)
plt.legend(['Max Temp', 'Min Temp'])
plt.xlabel('Month')
plt.ylabel('Temperature')   
plt.xticks(range(12), ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 
                       'Sep', 'Oct', 'Nov', 'Dec'], rotation=45)
plt.show()  
```

```python
# make a predictive model using historical data
# use the first 11 months of data to predict the last month
# use the average of the min and max temp as the prediction
import numpy as np
prediction = (data['Tmax'] + data['Tmin']) / 2
prediction = prediction[:334]
actual = data['Tmax']
actual = actual[334:]
plt.plot(actual)
plt.plot(prediction)
plt.legend(['Actual', 'Prediction'])
plt.xlabel('Day')
plt.ylabel('Temperature')
plt.show()
```

```python
# use sklearn linear regression to make a predictive model
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(np.arange(334).reshape(-1, 1), prediction)
prediction2 = model.predict(np.arange(334, 365).reshape(-1, 1))
plt.plot(actual)
plt.plot(prediction)
plt.plot(prediction2)
plt.legend(['Actual', 'Prediction', 'Prediction2'])
plt.xlabel('Day')
plt.ylabel('Temperature')
plt.show()
```
