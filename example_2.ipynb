```python
import pandas as pd
import matplotlib.pyplot as plt
# data = pd.read_csv('zodiac_languages.csv')
```

```python
import pandas as pd
import numpy as np

# Create an empty DataFrame with columns for zodiac signs and favorite programming languages
data = pd.DataFrame(columns=['Zodiac Sign', 'Favorite Language'])

# Generate random data for the DataFrame
zodiac_signs = ['Aries', 'Taurus', 'Gemini', 'Cancer', 'Leo', 'Virgo', 'Libra', 'Scorpio', 'Sagittarius', 'Capricorn', 'Aquarius', 'Pisces']
favorite_languages = ['Python', 'Java', 'JavaScript', 'C++', 'Ruby', 'Go', 'Rust', 'Swift', 'PHP', 'C#']

# Set the number of rows you want in the DataFrame
num_rows = 100

# Fill the DataFrame with random data
for _ in range(num_rows):
    zodiac = np.random.choice(zodiac_signs)
    language = np.random.choice(favorite_languages)
    data = data.append({'Zodiac Sign': zodiac, 'Favorite Language': language}, ignore_index=True)

# Print the DataFrame
print(data)
```

```python
print(data.head())
print(data.info())
```

```python
correlation = data.corr()
print(correlation)
```

```python
plt.figure(figsize=(10, 8))
plt.title('Correlation between Zodiac Signs and Favorite Programming Languages')
plt.imshow(correlation, cmap='coolwarm', vmin=-1, vmax=1)
plt.colorbar()
plt.xticks(range(len(data.columns)), data.columns, rotation=90)
plt.yticks(range(len(data.columns)), data.columns)
plt.show()
```

```python
import researchpy as rp

# Calculate Cramér's V for the categorical variables
categorical_vars = ['Zodiac Sign', 'Favorite Language']
cramer_v = rp.crosstab(data[categorical_vars[0]], data[categorical_vars[1]], prop='col', test='chi-square')['results'].iloc[2, 0]

# Print the Cramér's V value
print(f"Cramér's V: {cramer_v}")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Convert categorical variables into numerical labels
label_encoder = LabelEncoder()
for var in ['Zodiac Sign', 'Favorite Language']:
    data[var] = label_encoder.fit_transform(data[var])

# Calculate the correlation matrix
correlation = data.corr()

# Plot the correlation matrix as a heatmap
plt.figure(figsize=(10, 8))
plt.title('Correlation between Zodiac Signs and Favorite Programming Languages')
sns.heatmap(correlation, cmap='coolwarm', annot=True, vmin=-1, vmax=1)
plt.xticks(range(len(correlation.columns)), correlation.columns, rotation=90)
plt.yticks(range(len(correlation.columns)), correlation.columns)
plt.show()
```
